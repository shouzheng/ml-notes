正态分布 或 高斯分布

```python
# 标准的正态分布曲线可以了解下
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

x = np.arange(-3, 3, 0.001)
plt.plot(x, norm.pdf(x))  # norm.pdf为正态概率分布函数，pdf - probability density function
```

```python
# normal用于生成一组符合正态分布的数据，3个参数分别表示 数据的均值、标准差、数据的个数
values = np.random.normal(27000, 15000, 10000)
```

```python
# 使用可视化技术了解模拟数据的分布情况

import matplotlib.pyplot as plt

# hist方法用于绘制直方图，参数分别表示：数据、分组个数
plt.hist(values, 50)

# show方法将图形展示出来
plt.show()
```

均匀分布

```python
# uniform的参数：下限、上限、生成的数据量
values = np.random.uniform(-10.0, 10.0, 100000)

plt.hist(values, 50)
plt.show()
```

指数概率分布

```python
# 标准的指数概率分布曲线了解下
from scipy.stats import expon

x = np.arange(0, 10, 0.001)
plt.plot(x, expon.pdf(x))
```

```python
# 符合指数概率分布的模拟数据生成
values = np.random.exponential(1, 10000)
plt.hist(values, 50)
plt.show()
```

二项式概率质量函数

```python
# 二项式概率质量函数的标准图形了解下
from scipy.stats import binom

x = np.arange(0, 10, 0.001)
plt.plot(x, binom.pmf(x, 10, 0.5))  # n=10, p=0.5这两个参数会影响数据的形状
plt.show()
```

```python
plt.plot(x, binom.pmf(x, 8, 0.7))  # n, p这两个参数会影响数据的形状
plt.show()
```

```python
# 按照二项式概率质量函数的规律生成模拟数据
values = np.random.binomial(20, 0.5, 1000)
plt.hist(values, 50)
plt.show()
```

泊松概率质量函数

与正太分布非常像，但是有一点不同。如果你知道在给定时间段内某个事件发生的**平均次数**，那么就可以使用泊松概率质量函数来预测未来某个时间段内这个事情发生另一个次数的概率。

```python
# 标准的泊松概率质量图形了解下
from scipy.stats import poisson

mu = 500 # 均值，如：某网站平均每天访客500
x = np.arange(400, 600, 0.5)
plt.plot(x, poisson.pmf(x, mu))  # 根据图形，某一天有550个访客的概率为0.2%
```

```python
# 生成符合泊松分布的模拟数据
values = np.random.poisson(500, 365)
plt.hist(values, 50)
plt.show()
```
