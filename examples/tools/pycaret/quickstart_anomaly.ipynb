# [Anomaly Detection](https://pycaret.gitbook.io/docs/get-started/quickstart#anomaly-detection)

## Setup

```python
# load sample dataset
from pycaret.datasets import get_data
data = get_data('anomaly')
```

```python
from pycaret.anomaly import *
s = setup(data, session_id = 123)
```

## Create Model

```python
# functional API
iforest = create_model('iforest')
print(iforest)
```

```python
# functional API
models()
```

## Analyze Model

```python
# functional API
plot_model(iforest, plot = 'tsne')
```

```python
# functional API
plot_model(iforest, plot = 'umap')
```

```python
# functional API
evaluate_model(iforest)
```

## Assign Model

```python
# functional API
result = assign_model(iforest)
result.head()
```

## Predictions

```python
# functional API
predictions = predict_model(iforest, data = data)
predictions.head()
```

## Save the model

```python
# functional API
save_model(iforest, 'iforest_pipeline')
```

```python
# functional API
loaded_model = load_model('iforest_pipeline')
print(loaded_model)
```
