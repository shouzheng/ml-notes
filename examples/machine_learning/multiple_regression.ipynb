# 多元回归

```python
import pandas as pd

df = pd.read_excel('./data/cars.xls')
```

```python
df.head()
```

```python
df['Model_ord'] = pd.Categorical(df.Model).codes  # 数据处理 - 将类别型转换成数值型
X = df[['Mileage', 'Model_ord', 'Doors']]  # 特征筛选
y = df[['Price']]  # 定义标签

import statsmodels.api as sm
X1 = sm.add_constant(X)  # 添加一列常量特征，因为要表示成c+ax1+bx2的形式
est = sm.OLS(y, X1).fit()  # 最小二乘法 进行模型训练

est.summary()
```

```python
est.predict([[1.0, 8221, 10, 4], [1.0, 13196, 10, 4]])
```
