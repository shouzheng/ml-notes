# 获取数据

```python
import pandas as pd
df = pd.read_excel('./data/cars.xls')
df.head()
```

# 数据处理

在这里我们选择特征Mileage、Model、Doors，其中Model是字符型数据，需要进行类别编码。

```python
df['Model_ord'] = pd.Categorical(df.Model).codes
X = df[['Mileage', 'Model_ord', "Doors"]]
y = df['Price']

df.head()
```

# 建模

```python
# todo 关于这个包和算子，还有很多需要探索的地方

import statsmodels.api as sm
X1 = sm.add_constant(X)
est = sm.OLS(y, X1).fit()

est.summary()
```
